```python
dataset[dataset.isnull().any(axis=1)]
```

```python
#dataset[datadet.isnull().any(axis=1)] quantas linhas nulas?
dataset.dropna(inplace=True)

dataset
```

```python
x = dataset.iloc[:, 1:-1]
y = dataset.iloc[:, -1]

x
```

```python
x.dtypes
```

```python
x['smoking_status'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

cols = ['ever_married', 'Residence_type'] #unique para descobrir quantos rótulos únicos
le = LabelEncoder()
x[cols] = x[cols].apply(le.fit_transform)

x.head(10)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

#https://towardsdatascience.com/columntransformer-in-scikit-for-labelencoding-and-onehotencoding-in-machine-learning-c6255952731b
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), ['gender', 'work_type', 'smoking_status'])],
                       remainder='passthrough')

x = ct.fit_transform(x)

x[0]
```

```python
colunas = ['ohe1', 'ohe2', 'ohe3', 'ohe4', 'ohe5', 'ohe6', 'ohe7', 'ohe8', 'ohe9', 'ohe10', 'ohe11', 'ohe12', 'age', 'hypertension', 'heart_disease', 'ever_married', 'Residence_type', 'avg_glucose_level', 'bmi']

pd.DataFrame(x, columns=colunas)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

x_train
```
